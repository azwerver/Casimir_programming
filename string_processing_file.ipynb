```python
import string
random_string = 'Hello, this is a random string in which I put some words, but actually the string does not really say anything. But it is a really, really, really long string'

def count_words(your_string):
    """
    This function reads a string and returns the 
    amount of words in the string
    """
    words = your_string.split()

    return len(words)

    
def count_letters(your_string):
    """
    This function reads a string and returns the
    amount of letters that the string consists of
    """
    n = 0
    for letter in your_string:
        if letter in string.ascii_letters or letter in string.digits:
            n += 1
    
    return n

def count_tekens(your_string):
    """
    This function reads a string and returns the
    amount of tekens that the string consists of
    """
    list_of_tekens = string.ascii_letters + string.digits + string.punctuation
    n = 0
    for letter in your_string:
        if letter in list_of_tekens:
            n += 1
    
    return n

def find_occurance_of_word(your_string, word):
    """
    This function takes a string and a word and returns
    the amount of times that the word occurs in the string
    """
    words = your_string.split()
    count = words.count(word)
    
    # Ensure also words with punctuation are taken into account
    for teken in string.punctuation:
        new_word = word + teken
        count += words.count(new_word)
        
    return count


def find_occurance_of_all_words(your_string):
    """
    This function counts the occurance of all the words in a string
    """
    occurance_dict = {}

    no_interpunction = [c for c in your_string if c not in string.punctuation]
    new_string = ''.join(no_interpunction)
    new_string = new_string.upper()
    
    words = new_string.split()
    for word in words:
        if word in occurance_dict.keys():
            occurance_dict[word] += 1
        else:
            occurance_dict[word] = 1
            
    return occurance_dict


def print_word_plus_surroundings(your_string, word):
    """
    Some explanation
    """
    # Ensure that words with comma's, capitals etc are also taken into account
    word = word.upper()
    no_interpunction = [c for c in your_string if c not in string.punctuation]
    new_string = ''.join(no_interpunction)
    new_string = new_string.upper()
    
    word_list = new_string.split()
    
    return_list = []
    indices = [i for i, j in enumerate(word_list) if j == word]
    
    for index in indices:
        begin_index = index - 2
        end_index = index + 3
        if index < 2:
            begin_index = 0
        if index > len(word_list)-3:
            end_index = len(word_list)-1
        
        
        word_list_all = your_string.split()
        print_part = word_list_all[begin_index:end_index]
        print_string = ' '.join(print_part)
        
        return_list.append(print_string)
        
    return return_list
        
    
    
    
    
    
    
    
    
    


#########################
# print("The amount of words in the string is %d words." %count_words(random_string))
# print("The amount of letters in the string is %d letters." %count_letters(random_string))
# print("The amount of tekens in the string is %d tekens." %count_tekens(random_string))

# find_occurance_of_word(random_string, 'is')
# find_occurance_of_all_words(random_string)
print_word_plus_surroundings(random_string, 'Hello')
```
